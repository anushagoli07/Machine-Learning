```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.random.normal(200,10,500)
x.mean()
```

```python
x.std()
```

```python
len(x)
```

```python
plt.hist(x,bins=100)
plt.show()
```

```python
y=0.5*x+np.random.normal(10,5,500)
print(y.mean())
print(y.std())
```

```python
data=pd.DataFrame()
data['x']=x
data['y']=y
```

```python
z=np.column_stack([x,y])
z
```

```python
plt.scatter(z[:,0],z[:,1])
plt.show()
```

```python
plt.scatter(data['x'],data['y'])
plt.show()
```

```python
cov_mat=np.cov(z,rowvar=False)
cov_mat
```

```python
from numpy.linalg import eig
eig_val,eig_vec=eig(cov_mat)
print(eig_val)
print(eig_vec)
```

```python
cov_mat*eig_vec[0]
```

```python
eig_val[0]*eig_vec[0]
```

```python
eig_val/sum(eig_val)
```

```python
new_data=np.dot(z,eig_vec)
new_data=pd.DataFrame(new_data.real,columns=['pc1','pc2'])
new_data.head()
```

```python
cov_mat=np.cov(new_data,rowvar=False)
cov_mat
```

```python
plt.scatter(new_data['pc1'],new_data['pc2'])
plt.scatter(data['x'],data['y'])
plt.show()
```
